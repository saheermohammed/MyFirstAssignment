# 3. Ploting the bar graph of male vs female buyers participated in the sales.

```python
sns.set(font_scale=1.4,)
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(8, 8), rot=0,color=['red','blue'])
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title("Male vs Female Buyers", y=1.02)
```

# Insights

1.Female buyers are slightly higher than male buyers

# 4. Finding the top 5 cars based on their sales price.
